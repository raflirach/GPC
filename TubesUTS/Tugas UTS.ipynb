```python
def garis(x0,x1,y0,y1):
    x = []
    y = []
    x.append(x0)
    y.append(y0)
    deltaX = x1-x0
    deltaY = y1-y0
    if abs(deltaX) > abs(deltaY):
      step = abs(deltaX)
    else:
      step = abs(deltaY)
    x_inc = abs(deltaX) / step
    y_inc = abs(deltaY) / step
    for i in range(step):
        if x0<x1 :
            x0 = x0 + x_inc
            x.append(x0)
        else :
            x0 = x0 - x_inc
            x.append(x0)
        if y0<y1 :    
            y0 = y0 + y_inc
            y.append(y0)
        else :
            y0 = y0 - y_inc
            y.append(y0)
    return x,y
```
